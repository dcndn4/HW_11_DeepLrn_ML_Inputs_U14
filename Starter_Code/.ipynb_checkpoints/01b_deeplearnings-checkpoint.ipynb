# Deep Learning

This notebook shows how to build a deep learning model to predict the quality score of different wines

```python
# Initial imports
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

%matplotlib inline
```

<font color='red'>@@@</font>

Note: the csv is not comma seperated. "comma seperated values" can actually be "whatever seperated values". Here, it's **";" seperated.**

Q) What is our y label?

<font color='red'>@@@</font>

```python
# Read in data
filepath="C:/Users/CS_Knit_tinK_SC/Documents/My Data Sources/121421/01_winequality.csv"
#data = Path("../Resources/winequality.csv")
data = Path(filepath)
df = pd.read_csv(data, delimiter=";")
df.head()
```

```python
sorted(df["quality"].unique())
```

<font color='red'>@@@</font>

This seems like the easiest way to split x and y. No dropping columns, no fancy whatevers.

<font color='red'>@@@</font>

```python
# Create the features (X) and target (y) sets
X = df.iloc[:, 0:11].values
y = df["quality"].values
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
X = scaler.transform(X)
```

<font color='red'>@@@</font>

Regarding `# Define the model - shallow neural net`: 

This is a single layer network, aka, shallow net.

<font color='red'>@@@</font>

```python
# Define the model - shallow neural net
number_hidden_nodes = 8
number_input_features = 11

nn = Sequential()
# Hidden layer
nn.add(
    Dense(units=number_hidden_nodes, input_dim=number_input_features, activation="relu")
)
# Output layer
nn.add(Dense(units=1, activation="linear"))
```

<font color='red'>@@@</font>

What does mean_squared_error do?

<font color='red'>@@@</font>

```python
# Compile the model
nn.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse"])

# Train the model
model_1 = nn.fit(X, y, validation_split=0.3, epochs=200)
```

```python
# Plot the train and test loss function
plt.plot(model_1.history["loss"])
plt.title("loss_function - 1 hidden layer")
plt.legend(["loss"])
plt.show()
```

<font color='red'>@@@</font>

**Model 1:**

layer 1 = 8 neurons

**Model 2:**

layer 1 = 8 neurons

layer 2 = 4 neurons

<font color='red'>@@@</font>

```python
# Define the model - deep neural net
number_input_features = 11
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 4

nn = Sequential()
# First hidden layer
nn.add(
    Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(Dense(units=1, activation="linear"))
```

```python
# Compile model
nn.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse"])

# Fit the model
model_2 = nn.fit(X, y, validation_split=0.3, epochs=200)
```

```python
plt.plot(model_1.history["loss"])
plt.plot(model_2.history["loss"])
plt.title("loss_function - Training")
plt.legend(["1 hidden layer", "2 hidden layers"])
plt.show()
```

```python
# Train vs test for shallow net
plt.plot(model_1.history["loss"])
plt.plot(model_1.history["val_loss"])
plt.title("loss_function - Training Vs. Validation - 1 hidden layer")
plt.legend(["train", "test"])
plt.show()
```

```python
# Train vs test for deep net
plt.plot(model_2.history["loss"])
plt.plot(model_2.history["val_loss"])
plt.title("loss_function - Training Vs. Validation - 2 hidden layers")
plt.legend(["train", "test"])
plt.show()
```
